```python
import matplotlib.pyplot as plt

# Data
x = [10, 20, 30, 40, 50, 60, 70]

# Box Plot with all props including mean
plt.boxplot(
    x,
    sym='r+',  # outliers ka symbol
    showmeans=True,  # ✅ Mean ko show karega
    capprops=dict(color="blue", linewidth=2),   # whisker caps
    boxprops=dict(color="green", linewidth=2, linestyle="--"),  # box
    medianprops=dict(color="red", linewidth=2),  # median line
    whiskerprops=dict(color="orange", linewidth=2),  # whiskers
    flierprops=dict(marker='o', markerfacecolor='none',
                    markeredgecolor="purple", markersize=8),  # outliers
    meanprops=dict(marker="D", markerfacecolor="yellow",
                   markeredgecolor="black", markersize=10)  # ✅ mean marker style
)

plt.title("Customized Box Plot with Mean")
plt.xlabel("Values")
plt.show()
```
