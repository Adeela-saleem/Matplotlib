```python
import matplotlib.pyplot as plt
```

```python
#line plot
X=[1,2,3,4,5]
y=[3,4,2,5,6]
plt.plot(X,y)
plt.show()
```

# **Bar** **Plot**

---



```python
#bar plot
X=[1,2,3,4,5]
y=[3,4,2,5,6]
c=['red','blue','green','yellow','orange']
plt.bar(X,y,color=c)
plt.show()
```

```python
Subject=['python','java','assembly']
rating=[85,80,70]
plt.xlabel('Subject',fontsize=10)
plt.ylabel('Rating')
plt.title('Subject Rating')
plt.bar(Subject,rating,width=0.4,color='yellow',align='edge')#by default center mei likhy hoty
plt.show()
```

```python
#want to change edge color
plt.bar(Subject,rating,width=0.4,color='yellow',edgecolor='r',linewidth=5,linestyle=':',alpha=0.4)
plt.show()
```

```python
#make label

plt.bar(Subject,rating,width=0.4,color='yellow',label='subject rating')
plt.legend()#or legend mei bh direct label dy sakty
plt.show()
```

```python
#overlap side by side nhi hai
Subject=['python','java','assembly']
rating=[85,80,70]
z=[20,30,40]
plt.xlabel('Subject',fontsize=10)
plt.ylabel('Rating')
plt.title('Subject Rating')
plt.bar(Subject,rating,width=0.4,color='yellow',label='popularity1')
plt.bar(Subject,z,width=0.4,color='red',label='popularity2')
plt.legend()

plt.show()
```

```python
#double bar chart
import numpy as np

#  data
categories = ["A", "B", "C", "D"]
values1 = [4, 7, 3, 8]
values2 = [5, 6, 4, 7]
# X axis ke liye index
x = np.arange(len(categories))
width = 0.35  # bar ki chaurai
# Plot bars
plt.bar(x - width/2, values1, width,color='r', label="Group 1")
plt.bar(x + width/2, values2, width, label="Group 2")
# Labels aur title
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Double Bar Chart")
plt.xticks(x, categories,rotation=20)
plt.legend()
# Show plot
plt.show()
```

```python

import numpy as np

#  data
categories = ["A", "B", "C", "D"]
values1 = [4, 7, 3, 8]
values2 = [5, 6, 4, 7]
# X axis ke liye index
x = np.arange(len(categories))
width = 0.35  # bar ki chaurai
# Plot bars
plt.barh(x - width/2, values1, width,color='r', label="Group 1")
plt.barh(x + width/2, values2, width, label="Group 2")
# Labels aur title
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Double Bar Chart")
plt.xticks(x, categories,rotation=20)
plt.legend()
# Show plot
plt.show()
```
