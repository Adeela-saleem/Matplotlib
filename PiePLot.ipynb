```python
import matplotlib.pyplot as plt

# Data
labels = ["Apples", "Bananas", "Cherries", "Dates"]
sizes = [25, 30, 20, 25]  # percentages ya proportions

# Plot
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# Title
plt.title("Fruit Distribution")

# Show
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ["Apples", "Bananas", "Cherries", "Dates"]
sizes = [25, 30, 20, 25]
colors = ["red", "yellow", "pink", "orange"]

plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct="%1.1f%%",
    startangle=90,
    wedgeprops={
        'edgecolor': 'black',   # border ka rang
        'linewidth': 2,         # border ki motai
        'linestyle': '--',      # dashed line border
        'width': 0.7         # donut chart banane ke liye,

    }
)

plt.title("Fruit Distribution with Wedgeprops", pad=10)
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ["Apples", "Bananas", "Cherries", "Dates"]
sizes = [25, 30, 20, 25]
colors = ["red", "yellow", "pink", "orange"]

plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct="%1.1f%%",
    startangle=90,
    center=(2,3),   #  pie ka center (2,2) par shift kar diya
    wedgeprops={
        'edgecolor': 'black',   # border ka rang
        'linewidth': 2,         # border ki motai
        'linestyle': '--',      # dashed line border
        'width': 0.7 ,           # donut chart banane ke liye
    },
      rotatelabels=0.5

)

plt.title("Fruit Distribution with Wedgeprops & Center", pad=10)
plt.legend(loc=2)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data 1
labels1 = ["Apples", "Bananas", "Cherries", "Dates"]
sizes1 = [25, 30, 20, 25]
colors1 = ["red", "yellow", "pink", "orange"]

# Data 2
labels2 = ["Mango", "Orange", "Grapes", "Peach"]
sizes2 = [40, 20, 25, 15]
colors2 = ["gold", "darkorange", "purple", "lightgreen"]

# Pehla pie (left side)
plt.pie(
    sizes1,
    labels=labels1,
    colors=colors1,
    autopct="%1.1f%%",
    startangle=90,
    center=(0,0),   # center pe hi hai
    wedgeprops={'width':0.6, 'edgecolor':'black'}
)

# Dusra pie (right side, shift with center)
plt.pie(
    sizes2,
    labels=labels2,
    colors=colors2,
    autopct="%1.1f%%",
    startangle=90,
    center=(3,0),   #  isko right me shift kiya
    wedgeprops={'width':0.6, 'edgecolor':'black'}
)

plt.title("Double Pie Charts Example", pad=20)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data 1 (outer pie)
labels1 = ["Apples", "Bananas", "Cherries", "Dates"]
sizes1 = [25, 30, 20, 25]
colors1 = ["red", "yellow", "pink", "orange"]

# Data 2 (inner pie)
labels2 = ["Mango", "Orange", "Grapes", "Peach"]
sizes2 = [40, 20, 25, 15]
colors2 = ["gold", "darkorange", "purple", "lightgreen"]

# Outer pie
plt.pie(
    sizes1,
    labels=labels1,
    colors=colors1,
    autopct="%1.1f%%",
    startangle=90,
    radius=1.0,                 # badi circle
    wedgeprops={'width':0.3, 'edgecolor':'black'}
)

# Inner pie (overlap on same center)
plt.pie(
    sizes2,
    labels=labels2,
    colors=colors2,
    autopct="%1.1f%%",
    startangle=90,
    radius=0.7,                 # chhoti circle andar
    wedgeprops={'width':0.3, 'edgecolor':'black'}
)

plt.title("Overlapping Double Pie Chart", pad=20)
plt.show()
```
