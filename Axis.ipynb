```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, marker="o")

# Axis set karna
plt.axis([0, 6, 0, 12])  # [xmin, xmax, ymin, ymax]

plt.title("Axis ")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
import matplotlib.pyplot as plt

# Main plot
plt.plot([0,1,2,3], [0,1,4,9], label="Main Plot")

# Chhota plot figure ke andar (inset)
plt.axes([0.5, 0.5, 0.3, 0.3])  # [left, bottom, width, height]
plt.plot([0,1,2,3], [9,4,1,0], color="red")
plt.show()
```

Text

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x, y, marker="o")

# Simple text
plt.text(2, 7, "Important Point", color="red", fontsize=12)

plt.title("Text on Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.plot(x, y, marker="o")

# Annotate a point
plt.annotate(
    "Max Value",        # text
    xy=(5, 10),         # arrow point (x, y)
    xytext=(3, 8),      # text position
    arrowprops=dict(facecolor='black', shrink=0.05)
)

plt.title("Annotation Example")
plt.show()
```

```python
y2 = [1,3,7,6,9]

plt.plot(x, y, label="Line 1", marker="o")
plt.plot(x, y2, label="Line 2", marker="s")

# Annotate line 1 max
plt.annotate("Max L1", xy=(5,10), xytext=(3,11),
             arrowprops=dict(facecolor='blue', shrink=0.05))

# Annotate line 2 max
plt.annotate("Max L2", xy=(5,9), xytext=(3,10),
             arrowprops=dict(facecolor='green', shrink=0.05))

plt.title("Multiple Annotations")
plt.legend()
plt.show()
```
