tacked Area Plot:

Series 1 neeche hoti hai, Series 2 uske upar build hoti hai.

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y1 = [3, 5, 7, 9, 11]
y2 = [2, 4, 6, 8, 10]

# Stacked Area Plot
plt.stackplot(x, y1, y2, labels=["Series 1", "Series 2"], colors=["skyblue", "orange"], alpha=0.7)

plt.title("Stacked Area Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y1 = [3, 5, 7, 9, 11]
y2 = [2, 4, 6, 8, 10]

# Stacked Area Plot
plt.stackplot(x, y1, y2, labels=["Series 1", "Series 2"], colors=["skyblue", "orange"],baseline='sym')

plt.title("Stacked Area Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```
