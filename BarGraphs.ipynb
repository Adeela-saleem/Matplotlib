```python
import matplotlib.pyplot as plt

# Data
subjects = ["Math", "Computer", "Physics", "English", "Chemistry"]
ratings = [8, 9, 7, 6, 7]

# Plot
plt.bar(subjects, ratings,width=0.4)
plt.xlabel("Subjects",fontsize=15)
plt.ylabel("Ratings (out of 10)")
plt.title("Subject Ratings")
plt.show()
```

```python
#apply label
import matplotlib.pyplot as plt

# Data
subjects = ["Math", "Computer", "Physics", "English", "Chemistry"]
ratings = [8, 9, 7, 6, 7]

# Plot
plt.bar(subjects, ratings,width=0.4,label='rating')
plt.xlabel("Subjects",fontsize=15)
plt.ylabel("Ratings (out of 10)")
plt.title("Subject Ratings")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Subjects
subjects = ["Math", "Physics", "Chemistry", "CS", "English"]

# Ratings from 3 different students
student1 = [8, 7, 9, 8, 6]
student2 = [6, 8, 7, 9, 7]
student3 = [7, 6, 8, 7, 8]

x = np.arange(len(subjects))  # X-axis positions
width = 0.25  # bar ki width

plt.bar(x - width, student1, width, label="Student 1")
plt.bar(x, student2, width, label="Student 2")
plt.bar(x + width, student3, width, label="Student 3")

plt.xlabel("Subjects")
plt.ylabel("Ratings (out of 10)")
plt.title("Subject Ratings by Students")
plt.xticks(x, subjects,rotation=10)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Subjects
subjects = ["Math", "Physics", "Chemistry", "CS", "English"]

# Ratings from 3 different students
student1 = [8, 7, 9, 8, 6]
student2 = [6, 8, 7, 9, 7]
student3 = [7, 6, 8, 7, 8]

x = np.arange(len(subjects))  # X-axis positions
width = 0.25  # bar ki width

plt.barh(x - width, student1, width, label="Student 1")
plt.barh(x, student2, width, label="Student 2")
plt.barh(x + width, student3, width, label="Student 3")

plt.xlabel("Subjects")
plt.ylabel("Ratings (out of 10)")
plt.title("Subject Ratings by Students")
plt.xticks(x, subjects,rotation=10)
plt.legend()
plt.show()
```
