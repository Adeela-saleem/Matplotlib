Stem Plot Kya Hai?

Definition: Stem plot ek aisa graph hai jo data ko discrete vertical lines (stems) ke through show karta hai.

Kaise hota hai?

Har data point ke liye ek line (stem) x-axis se lekar us value tak draw hoti hai.

Line ke upar ek marker (dot, circle, square) hota hai jo exact value show karta hai.

ðŸ‘‰ Isse data ka distribution aur spikes easily samajh aata hai.

ðŸ”¹ Kab Use Hota Hai?

Jab aapko discrete data points highlight karne hain.

Time series data ka sample points show karna.

Signal processing me (jaise sine/cosine values).

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 2*np.pi, 8)   # 8 points 0 to 2Ï€ tak
y = np.sin(x)

# Stem plot
plt.stem(x, y)

plt.title("Basic Stem Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 2*np.pi, 8)
y = np.cos(x)

# Stem plot with style
plt.stem(
    x, y,
    linefmt='r--',    # line ka style (r = red, -- = dashed)
    markerfmt='go',   # marker style (g = green, o = circle)
    basefmt='k-'      # base line (x-axis) ka style (k = black)
)

plt.title("Styled Stem Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
