Normally plt.plot() ek hi figure (graph) banata hai.

Subplots ka use tab hota hai jab aap ek hi window/figure me multiple graphs dikhana chahte ho.

Matlab ek hi canvas ko chhote chhote grids (rows Ã— columns) me tod dete hain, aur har grid me ek graph hota hai.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [25, 16, 9, 4, 1]

# 1st subplot
plt.subplot(1, 2, 1)   # (rows=1, cols=2, position=1)
plt.plot(x, y1, color="blue")
plt.title("Graph 1")

# 2nd subplot
plt.subplot(1, 2, 2)   # (rows=1, cols=2, position=2)
plt.plot(x, y2, color="red")
plt.title("Graph 2")

plt.tight_layout()  # spacing adjust
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.subplot(2, 2, 1)
plt.plot(x, y, color="blue")
plt.title("Plot 1")

plt.subplot(2, 2, 2)
plt.bar(x, y, color="orange")
plt.title("Plot 2")

plt.subplot(2, 2, 3)
plt.scatter(x, y, color="green")
plt.title("Plot 3")

plt.subplot(2, 2, 4)
plt.hist(y, color="purple")
plt.title("Plot 4")

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 6))

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

axs[0, 0].plot(x, y, color="blue")
axs[0, 0].set_title("Line Plot")

axs[0, 1].bar(x, y, color="orange")
axs[0, 1].set_title("Bar Plot")

axs[1, 0].scatter(x, y, color="green")
axs[1, 0].set_title("Scatter Plot")

axs[1, 1].hist(y, color="purple")
axs[1, 1].set_title("Histogram")

plt.tight_layout()
plt.show()
```
