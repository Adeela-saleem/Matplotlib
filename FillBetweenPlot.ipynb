fill_between() Plot kya hai?

Ye function ek line ke neeche ka area fill karta hai.

Ye area plot banane ka base function hai.

Iska use mostly hota hai jab aapko kisi range ya difference ko highlight karna ho.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 6, 8, 10]

plt.plot(x, y, color="blue")  # line
plt.fill_between(x, y, color="skyblue", alpha=0.5)  # area fill

plt.title("Fill Between Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin(x)", color="blue")
plt.plot(x, y2, label="cos(x)", color="red")

# Difference ka area fill
plt.fill_between(x, y1, y2, where=(y1 > y2), color="green", alpha=0.3, label="sin > cos")
plt.fill_between(x, y1, y2, where=(y1 <= y2), color="orange", alpha=0.3, label="cos ≥ sin")

plt.title("Fill Between Two Lines")
plt.legend()
plt.show()
```
