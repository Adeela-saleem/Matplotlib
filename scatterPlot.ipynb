```python
import matplotlib.pyplot as plt
```

```python
X=[1,2,3,4,5,6,7]
y=[2,4,5,6,3,7,8]
colors=['r','b','g','y','r','g','r']
plt.scatter(X,y,c=colors)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#dots sizzes differece + trasperancy
X=[1,2,3,4,5,6,7]
y=[2,4,5,6,3,7,8]
sizes=[100,200,300,400,500,600,700]
plt.scatter(X,y,c='r',s=sizes,alpha=0.5)  #between 0,1
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#change marker
X=[1,2,3,4,5,6,7]
y=[2,4,5,6,3,7,8]
plt.scatter(X,y,c='r',alpha=0.5,marker='*',s=100,edgecolor='g',linewidth=3)  #between 0,1
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#color change by cmap
#change marker
X=[1,2,3,4,5,6,7]
y=[2,4,5,6,3,7,8]
colors=[10,20,30,50,70,60,80]  #bertween 0 to 100
plt.scatter(X,y,s=100,c=colors,cmap='BrBG')
t=plt.colorbar()
t.set_label('Color Bar')
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#add 2 graphs

X=[1,2,3,4,5,6,7]
y=[2,4,5,6,3,7,8]
z=[7,5,6,8,9,3,4]
colors=[10,20,30,50,70,60,80]
plt.scatter(X,y,s=100,c=colors,cmap='BrBG')
plt.scatter(X,z,s=100,c='r')
t=plt.colorbar()
t.set_label('Color Bar')
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
