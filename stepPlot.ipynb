Step Plot kya hai?

Ye ek special type of line plot hai jisme data points ko straight lines ki jagah steps ki form mein connect kiya jata hai.

Matlab har value ek range tak constant rahti hai, phir ekdum jump hoti hai.

ðŸ”¹ Kab use hota hai?

Jab aapko discrete changes dikhane ho instead of smooth trend.

Jaise:

Electricity billing slabs

Population growth in intervals

Digital signals (0 aur 1)

```python
import matplotlib.pyplot as plt

# Data
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 0, 2, 3, 2]

# Step Plot
plt.step(x, y, where="mid", color="blue", linewidth=2, label="Step Plot")

plt.title("Simple Step Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 0, 2, 3, 2]

plt.step(x, y, where="pre", label="pre", color="red")
plt.step(x, y, where="mid", label="mid", color="green")
plt.step(x, y, where="post", label="post", color="blue")

plt.title("Step Plot with 'where' options")
plt.legend()
plt.show()
```
